```python
# IntÃ©rprÃ©tation statistique:
inertia=[]

for n_clusters in range(2,11):
    km = KMeans(n_clusters=n_clusters, random_state=42)
    km.fit(df_minmax)
    inertia.append(km.inertia_)

plt.plot(range(2,11), inertia)
```
